```python
responses = pd.read_csv('../Niteo/Emergency_Response_Incidents.csv')
```

```python
responses = responses.rename(columns={'Creation Date':'SDate','Closed Date':'EDate','Incident Type':'Incidents'})
```

```python
def parseSt(x):
    if '&' in x:
        return x
    else:
        x = x.split(' ')
        if len(x)==3:
            return x[1]
        elif len(x) == 4:
            y = x[1]+x[2]
            return y
test['Street'] = test.Location.apply(parseSt)
```
