```python
sns.boxplot(x = contestants.Season, y = contestants.Age, data = contestants)
```

```python
def state(x):
    x = x.split(', ')
    return x[1]
contestants['State'] = contestants.Hometown.apply(state)

plt.figure(figsize=(20,10))
plt.xticks(rotation = 45)
sns.countplot(contestants.State)
```
