```python
def repTitle(x):
    if 'debt' in x:
        x='debt'
    return x
df['title']=df.title.apply(repTitle)
df.title.value_counts()
```

```python
def defCred(x):
    if 'credit card' in x:
        if 'consolidat' in x:
            x='credit card consolidation'
        else:
            x='credit card debt'
    return x
df['title']=df.title.apply(defCred)
df.title.value_counts()
```

```python
def WedDef(x):
    if 'wedding' in x:
        x='wedding'
    return x
df['title']=df.title.apply(WedDef)
df.title.value_counts()
```

```python
def MedDef(x):
    if 'medical' in x:
        x='medical'
    return x
df['title']=df.title.apply(MedDef)
df.title.value_counts()
```

```python
def PerDef(x):
    if 'personal' in x:
        x='personal'
    return x
df['title']=df.title.apply(PerDef)
df.title.value_counts()
```

```python
def HomeDef(x):
    if 'home' in x:
        x='home improvement'
    return x
df['title']=df.title.apply(HomeDef)
df.title.value_counts()
```
